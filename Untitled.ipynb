```python
import os
import shutil
from tqdm import tqdm
path_old_img="../../../../../home/data/Liver_Vessel_Segmentation/colorectal/data_corrected_2d/images_lv_2d"
path_old_msk="../../../../../home/data/Liver_Vessel_Segmentation/colorectal/data_corrected_2d/masks_vs_2d"
path_new="../../../vssSeg/Data/colorectal_fold"
for i in tqdm(range(1,6)):
    name_folder=f"fold_{i}"
    
    path_fold=os.path.join(path_new,name_folder)
    
    path_fold_train=os.path.join(path_fold,"train")
    path_fold_train_img=os.path.join(path_fold_train,"images")
    path_fold_train_msk=os.path.join(path_fold_train,"masks")
    
    path_fold_val=os.path.join(path_fold,"val")
    path_fold_val_img=os.path.join(path_fold_val,"images")   ##
    path_fold_val_msk=os.path.join(path_fold_val,"masks")    ##
    
    os.mkdir(path_fold)
    os.mkdir(path_fold_train)
    os.mkdir(path_fold_train_img)
    os.mkdir(path_fold_train_msk)
    os.mkdir(path_fold_val)
    os.mkdir(path_fold_val_img)
    os.mkdir(path_fold_val_msk)
    
    folds=[1,2,3,4,5]
    number=folds.pop(i-1)
    
    files_val=fold_sep[number]
    
    for file in files_val:
        file_img_val_old_name=os.path.join(path_old_img,file)
        file_img_val_new_name=os.path.join(path_fold_val_img,file)
        
        file_msk_val_old_name=os.path.join(path_old_msk,file)
        file_msk_val_new_name=os.path.join(path_fold_val_msk,file)
        
        shutil.copy(file_img_val_old_name, file_img_val_new_name)
        shutil.copy(file_msk_val_old_name, file_msk_val_new_name)

    files_train=np.concatenate([fold_sep[i] for i in folds ])
    
    for file in files_train:
        file_img_train_old_name=os.path.join(path_old_img,file)
        file_img_train_new_name=os.path.join(path_fold_train_img,file)
        
        file_msk_train_old_name=os.path.join(path_old_msk,file)
        file_msk_train_new_name=os.path.join(path_fold_train_msk,file)
        
        shutil.copy(file_img_train_old_name, file_img_train_new_name)
        shutil.copy(file_msk_train_old_name, file_msk_train_new_name)
```
